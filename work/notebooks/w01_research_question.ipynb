## 2. The question: decision, action, cost of a wrong call

- **Decision to improve:** This work improves the allocation of editorial resources by deciding which specific pages should be queued for a content refresh.
- **Who acts on it:** Content managers, SEO specialists, and copywriters.
- **Action taken:** They will review the top-ranked pages, rewrite sections, update obsolete information, or improve user engagement elements.
- **Cost of a wrong call:** A false positive (ranking a stable page highly) wastes editorial time and budget on content that didn't need fixing. A false negative (missing a truly declining page) leads to a continuous loss of search visibility and traffic, which damages revenue.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os, sys, subprocess
import pandas as pd

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

# سحب البيانات وإعداد بيئة العمل
if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
elif "work" in os.getcwd() or "notebooks" in os.getcwd():
    # للرجوع للمجلد الرئيسي إذا كنت تعمل محلياً
    while not os.path.exists("data"):
        os.chdir("..")

print("data is Downloaded :", os.getcwd())

# الآن سيتمكن الكود من إيجاد الملف وقراءته بنجاح
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# 1. Total number of pages available for analysis
total_pages = df.shape[0]

# 2. Percentage of pages that are currently declining
df["is_declining"] = df["trend_direction"].str.lower().eq("down").astype(int)
decline_rate = df["is_declining"].mean() * 100

# 3. High visibility pages (impressions >= 500) that are stale (not updated in >= 180 days)
stale_and_visible = df[(df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)].shape[0]

print(f"Real Number 1 (Total Inventory Scale): {total_pages} unique pages.")
print(f"Real Number 2 (Prevalence of Decline): {decline_rate:.1f}% of pages are in a 'down' trend.")
print(f"Real Number 3 (Immediate Opportunity): {stale_and_visible} pages are highly visible but haven't been updated in over 180 days.")
```

## 4. Careful words: what I can and can't claim
- **What I CAN claim:** This system provides decision-support by ranking pages based on observed historical patterns and measured signals (like impressions, position, and age). It highlights directional decay risks to help human reviewers spend their limited time more efficiently.
- **What I CANNOT claim:** This model does not prove causal relationships (e.g., I cannot claim that an old page *caused* the traffic drop, nor that updating it *guarantees* recovery). Furthermore, this is not a system for 'predicting Google's algorithm'—it only models patterns within our internal dataset to flag optimization opportunities.
